```python
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import urllib3
from bs4 import BeautifulSoup
import csv
import html
```

```python
def get_link(data):
    # data = open('./HomeSource12.html','r')
    posts_link = []
    soup = BeautifulSoup(data, 'html.parser')
    for post in soup.find_all('div', {'class': 'post-card-item-af972'}):
        posts_link.append(post.find('a').get('href'))

    return posts_link
```

```python
chromrdriver = "./chromedriver"
os.environ["webdriver.chrome.driver"] = chromrdriver
driver = webdriver.Chrome(chromrdriver)
driver.get("https://divar.ir/s/tehran")


ScrollNumber = 30
posts_link = []
for i in range(1,ScrollNumber):
    driver.execute_script("window.scrollTo(1,50000)")
    posts_link.append(get_link(driver.page_source))
    print(i)
    time.sleep(5)

# file = open('HomeSource12.html', 'w')
# file.write(driver.page_source)
# file.close()

driver.close()
```

```python
flat_list = [item for sublist in posts_link for item in sublist]
flat_list
```
